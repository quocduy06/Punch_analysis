```python
import pandas as pd
import numpy as np
import glob
import os

# Cấu hình đường dẫn và ánh xạ nhãn cường độ
data_dir = r"C:\Users\Admin\Downloads\draft_one_signal_final_project\src"
intensity_map = {
    "LIGHT": 1,
    "MEDIUM": 2,
    "INTENSE": 3,
    "FATIGUE": 3
}

# Tìm tất cả file CSV thô (loại trừ file master cũ)
all_files = [f for f in glob.glob(os.path.join(data_dir, "*.csv")) if "master" not in f]
print(f"✅ Đã tìm thấy {len(all_files)} file dữ liệu để xử lý.")
```

```python
master_list = []

for file_path in all_files:
    # 1. Đọc file
    df = pd.read_csv(file_path)
    file_name = os.path.basename(file_path)
    
    # 2. Bóc tách Metadata từ tên file (Ví dụ: INTENSE_CLEAN_07_timestamp.csv)
    parts = file_name.split('_')
    session, prop, trial_id = parts[0], parts[1], parts[2]
    
    # 3. Feature Engineering: Tính Magnitude
    # Công thức: sqrt(x^2 + y^2 + z^2)
    df['Acc_Mag'] = np.sqrt(df['AccX']**2 + df['AccY']**2 + df['AccZ']**2)
    
    # 4. Dán nhãn tự động cho Duy (Member 3)
    df['Intensity_Label'] = intensity_map.get(session, 0)
    df['Trial_ID'] = f"{session}_{trial_id}"
    df['Property'] = prop
    
    master_list.append(df)
    print(f"➡️ Đang xử lý: {file_name}")

print("\n✅ Hoàn tất sơ chế tất cả các file.")
```

```python
# Gộp và Lưu file Master
master_df = pd.concat(master_list, ignore_index=True)
master_output = os.path.join(PROCESSED_DIR, "master_dataset_aikido.csv")
master_df.to_csv(master_output, index=False)

print(f"🎊 THÀNH CÔNG! File Master đã nằm gọn trong: {master_output}")
```
